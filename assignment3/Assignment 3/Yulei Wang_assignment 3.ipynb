```python
import turtle
import datetime
import time

turtle.color("white")
turtle.screensize(8000,6000,"black")
turtle.speed(10)

for i in range(50000):

    #Extract time elements of second, minute, hour

    today = datetime.datetime.now()  
    s = today.second
    m = today.minute
    h = today.hour
    
#I created 3 turtles,they will extract current number of second, minuet, hour respectively from computer to draw the polygons.
#The number of this polydon edge same as current number of seconds.    
    turtle.circle(50,steps=s)
    
    
    turtle.setheading(0)
    turtle.penup()
    turtle.forward(200)
    turtle.pendown()
    
#The number of this polydon edge same as current number of minutes.      
    turtle.circle(50,steps=m)
    
    
    turtle.setheading(0)
    turtle.penup()
    turtle.forward(200)
    turtle.pendown()
    
#The number of this polydon edge same as current number of hours.      
    turtle.circle(50,steps=h)
    
    
    turtle.setheading(180)
    turtle.penup()
    turtle.forward(400)
    turtle.pendown()
    
    turtle.setheading(0)
    
    turtle.clear()

    
turtle.done()
```
