__**Data Cleaning**__

As mentioned above we need to clean the dataset and account for missing values, simplify variables to avoid overfitting, create additional variables and then remove unecessary information in order to accurately plot our findings and ues these to evidence our initial hypothesis. 

```python
# Finds how many missing values within the 'Cabin', 'Age' and 'Embarked' columns

train.isnull().sum() # Shows the amount of missing and unusable values within the dataset
```

```python
# Shows the missing values within the 'Embarked' column

train[train['Embarked'].isnull()]
```

```python
# Fills the missing 2 values within the 'Embarked' column

# Creates a variable called embarked mean being the mean value/location of embarkment for the dataset

embarkedmean = train.Embarked.mean 

# Fills the empty cells within the 'Embarked' column with the mean value/location

train['Embarked'] = train['Embarked'].fillna(embarkedmean) 
```

Now that the values for the 'Embarked' column have been filled we can move onto filling the gaps within the 'Age' and 'Cabin' column. However unlike for the 'Embarked' column I feel that a different approach is needed for these columns as they play a bigger factor in survival rate. The most logical way that I can think about assigning the missing values within the 'Age' column is looking at the mean and standard deviation and then generating random numbers between these two values and then using this as an age basis to assign the missing values. However for the 'Cabin' column I will choose to drop this as there are too many incomplete values within the dataset and attempting to fill this may cause inaccuracies, and this feature directly relates to class as the first letter corresponds to the deck; this meaning we would probably get similar results comparing variables against 'Pclass' or a completed 'Cabin' (Deck) column.

```python
# Shows the missing values within the 'Age' column

train[train['Age'].isnull()]
```

```python
# Gains the average, std and number of missing values within the 'Age' column within the dataset

averageage = train['Age'].mean()
stdage = train['Age'].std()
nanvaluesage = train['Age'].isnull().sum()

# Generates random numbers between (mean - std) and (mean + std) which can then be assigned to fill the gaps in the dataset

rand1 = np.random.randint(averageage - stdage, averageage + stdage, size = nanvaluesage)

# Shows graphically the change in age range when missing values have been filled

fig, (axis1,axis2) = plt.subplots(1,2,figsize=(12,4))
axis1.set_title('Original Age Range - Titanic Passengers')
axis2.set_title('New Age Range - Titanic Passengers')

# Plots original Age range minus the missing values

train['Age'].dropna().astype(int).hist(bins=60, ax=axis1)

# Fills the missing values within the 'Age' column with the random values as generate above

train['Age'][np.isnan(train['Age'])]= rand1

# Converts the float variable within 'Age' into an int

train['Age'] = train['Age'].astype(int)

# Shows graphically the change in age range when missing values have been filled

train['Age'].hist(bins=60, ax=axis2)
```

As you can see from the graphs above, the shape of the 'Age' graph is almost unchanged apart from the added values around the centre of the graph which are due to the filling of the missing values made by the random generator. (Another way in which I had invisioned filling in the missing values was by extracting the "Titles' and finding the average age per 'Title' and then using this to assign the missing values, however after several attempts I could not get this to work, so instead settled for an simple yet effective solution)

Next I emit the 'Cabin' column as explained above, due to the large amounts of incomplete data and for the fact that it wont make a difference when analysing the data. 

```python
# Emits the values within the 'Cabin' column

train = train.drop(['Cabin'], axis=1)
```

```python
# Shows the amount of missing and unusable values within the dataset

train.isnull().sum()
```

As shown above, all of the missing fields within the dataset have now been filled, this now means we can further sort the data such as by adding new columns for 'Family Size' and 'Title' so that we have some new variables to analyse when graphically displaying results. The way in which I have gone about further editing the data can be shown below.

```python
# Extracts the 'Title' for each person within the 'Name' column 

train['Title'] = train.Name.str.extract(' ([A-Za-z]+)\.', expand=False)

pd.crosstab(train['Title'], train['Sex']) # Displays the frequency of 'Titles' for each person on board
```

```python
# Obtains the Titles and Surnames from the 'Name' column (However with double-barrelled surnames it can miss the second half, e.g.'Vander Planke', displays as 'Vander')

def obtainsurnames(train):
    train['Surname'] = train.Name.apply(lambda x: x.split(' ')[0])
    return train

obtainsurnames(train)
```

```python
# Puts Titles into different categories based upon uniqueness 

# Replaces various titles with grouped headings

train['Title'] = train['Title'].replace(['Lady', 'Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Jonkheer', 'Dona'], 'Unique')    
train['Title'] = train['Title'].replace(['Countess', 'Lady', 'Sir'],'Royal')
train['Title'] = train['Title'].replace('Mlle', 'Miss')
train['Title'] = train['Title'].replace('Ms', 'Miss')
train['Title'] = train['Title'].replace('Mme', 'Mrs')
```

```python
# Now Titles and Surnames have been obtained from the 'Name' column we can now omit/drop this

train = train.drop(['Name'], axis=1)
```

```python
# Creates a new column called 'FamilySize' which is a summation of 'SibSp' and 'Parch'

train['FamilySize'] = train['SibSp'] + train['Parch'] + 1
```

```python
# Now FamilySize has been created we can omit/drop the 'SibSp' and 'Parch' columns

train = train.drop(['SibSp','Parch'], axis=1)
```

```python
# This will find the max/min values for the 'Fare' column (Useful to see how it will break up the fares into groups/bands)

train.Fare.min(), train.Fare.max()
```

```python
# Shows the first 10 values within the 'Fare' column, showing the ticket prices before groupings

train.Fare.head(10) 
```

```python
# This will add a mapping feature to the fares so that it can more easily be visualised into groups 

train['FareBand'] = pd.qcut(train['Fare'], 4, labels = [1, 2, 3, 4])
```

```python
# Shows the first 10 values within the new 'FareBand' column, showing a useful comparison against the 'Fare' column as you can see what price ranges were assigned to each band

train.FareBand.head(10)
```

```python
# Removes the 'Fare' column as the new 'FareBand column has now been created

train = train.drop(['Fare'], axis=1)
```
