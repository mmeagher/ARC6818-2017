The basic information

```python
df = pd.read_csv("train.csv")
df
```

The barchart shows the relationship between Embarked and survival chance
. It is clear that S contributed most of the death

```python
sns.set(style="whitegrid")  

ax = sns.countplot(x="Embarked", hue="Survived", data=df,palette=sns.color_palette("Set2", 8)) 
```

The linechart shows the relationship among Embarked, Sex, survival chance.It is clear that Female survived more

```python
sns.set(style="whitegrid") 
sns.pointplot(x="Embarked", y="Survived", hue="Sex", data=df,
              palette=sns.color_palette("Set2", 8),
              markers=["^", "o"], linestyles=["-", "--"]);
```

The barchart shows the relationship between Pclass and survival chance
. The death rate increased from class 1 to class 3

```python
sns.set(style="whitegrid")  

ax = sns.countplot(x="Pclass", hue="Survived", data=df,palette=sns.color_palette("Set2", 8)) 
```

This shows the survival rate decreased sharply in male from class 1 to class 2，that in female decreased sharply from class 2 to class 3

```python
sns.set(style="whitegrid")  
sns.pointplot(x="Pclass", y="Survived", hue="Sex", data=df,
              palette=sns.color_palette("Set2", 8),
              markers=["^", "o"], linestyles=["-", "--"]);
```

Define agegroups. Adults has most number of death,and child has high survival rate

```python
df["AgeRange"] = pd.cut(df["Age"], [0, 16, 65, 100], labels=["Child", "Adult", "Elderly"])
survived_age = df.groupby("AgeRange")["Survived"].mean()

sns.set(style="whitegrid")  
ax = sns.countplot(x="AgeRange", hue="Survived", data=df,palette=sns.color_palette("Set2", 8)) 
```

This chart shows male child survived about 43%, female child survived about 67%. The female adult has highest rate. It is around 77%, while no elderly female survived

```python
g = sns.FacetGrid(df,col="Sex", size=10, aspect=.3)
g.map(sns.barplot, "AgeRange", "Survived",palette=sns.color_palette("Set2", 8));
```

Calculate the name lenth，try to find out it is relevant for understanding survival or not

```python
df["Namelength"]=df.Name.apply(len)
df.sort_values("Namelength")
df
```

It shows male name usually shorter than female，but the relationship between name length and survival chance not very clear，because it has too much lenth of name

```python
sns.factorplot(x="Sex", y="Namelength", hue="Survived", data=df, kind="swarm",palette=sns.color_palette("Set2", 8),size=6);
```

So define NameRange groups，from this chart it is obvious that short name male survived less than middle lenth name male. While in female, the name lenth may not be a contributor to the death

```python
df["NameRange"] = pd.cut(df["Namelength"], [0, 25, 50, 85], labels=["ShortName", "MiddleName", "LongName"])
survived_age = df.groupby("Namelength")["Survived"].mean()
d = sns.FacetGrid(df,col="Sex", size=10, aspect=.3)
d.map(sns.barplot, "NameRange", "Survived",palette=sns.color_palette("Set2", 8));
```
