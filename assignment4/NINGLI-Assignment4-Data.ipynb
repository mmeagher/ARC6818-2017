#### After getting the number of Embarked, calculate the rate.

```python
def draw_pie(labels,quants):   
    plt.figure(1, figsize=(6,6))   
    expl = [0,0.1,0.1]      
    colors  = ["r","g","b"]     
    plt.pie(quants, explode=expl, colors=colors, labels=labels, autopct='%1.1f%%',pctdistance=0.8, shadow=0)  
    plt.title('Embarked Distribution',)  

labels   = ['S','C','Q']   
quants   = [644,168,77]   
draw_pie(labels,quants)  

plt.show()
```

#### Calculate survival number in different catagories

```python
pd.crosstab(df["Embarked"],df["Survived"]).plot(kind="bar")
plt.title('Embarked: Survived')

pd.crosstab(df["Sex"],df["Survived"]).plot(kind="bar")
plt.title('Sex: Survived')

pd.crosstab(df["Age"],df["Survived"]).plot(kind="bar")
plt.title('Age: Survived')
```

#### Distribution for each one in 2-dimensional diagram to show more info

```python
fig, ax = plt.subplots()

xlive = df.Age[df.Survived==1]
ylive = df.Fare[df.Survived==1]
ax.set_xlabel('Age')
ax.set_ylabel('fare')
m = ax.scatter(xlive, ylive, 5, 'r')
ax.hold

plt.title('Survived') 
plt.show()

fig, ay = plt.subplots()

xdie = df.Age[df.Survived==0]
ydie = df.Fare[df.Survived==0]
ay.set_xlabel('Age')
ay.set_ylabel('fare')
n = ay.scatter(xdie, ydie, 5, 'g')
ay.hold

plt.title('Died') 
plt.show()
```
