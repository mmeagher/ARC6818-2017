```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

df = pd.read_csv('train.csv',header=0)
```

```python
df
```

```python
df['Name_lengeh']=df.Name.apply(len)
sns.factorplot(x="Sex", y="Name_lengeh", hue="Survived", data=df, kind="swarm")
plt.show()
```

```python
#This part shows the relationship between the length of name and the survival rate of 2 genders.
```

FamilySize

```python
SS = pd.DataFrame()
fig = plt.figure()
ax = fig.add_subplot(111)


df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
SS['PassengerNum'] = df[['FamilySize','Survived']].groupby(['FamilySize'])['Survived'].count()
SS['FamSize'] = SS.index.values
SS['Survived'] = df[['FamilySize','Survived']].groupby(['FamilySize'])['Survived'].sum()
SS['Ratio'] = SS.Survived / SS.PassengerNum * 100

pd.crosstab(df["FamilySize"],df["Survived"]).plot(kind="line")
SS.Ratio.plot(kind='bar', color='red', ax=ax, label = 'Ratio')
ax.set_ylabel('%')
ax.legend(['Ratio'])
plt.show()
```

```python
#The FamilySize under 4 has the higher death rate.As those which have more than 4 have lower death rate.
```

Embarked

```python
sns.factorplot('Embarked','Survived', data=df,size=2,aspect=4)    

plt.show()
```

```python
#In Embarked C has the higher rate of survival
```

Age & Fare & Pclass

```python
fig, ax = plt.subplots()

ylive = df.Fare[df.Survived==1]
ydie = df.Fare[df.Survived==0]
xlive = df.Age[df.Survived==1]
xdie = df.Age[df.Survived==0]

m = ax.scatter(xlive, ylive, 5, 'pink')
ax.hold
n = ax.scatter(xdie, ydie, 5, 'red')
ax.hold

ax.set_xlabel('Age')
ax.set_ylabel('Fare')

ax.legend([m,n],['Survived','Died'])

plt.show()
```

```python
#This part is describe the figure for age and fare without the restriction of Pclass.From this part,it is clear to see that the price of tickets are focus on 0-100,the passengers' mostly from 15-50.And the dead rate is concentrite on this part,especially in 20-to-50 people.
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('Age group')
ax.set_ylabel('Fare group')
ax.set_zlabel('Pclass group')

a = df['Pclass']

for z1 in a:
    xlive= df.Age[(df.Survived==1) & (df.Pclass==z1)]
    ylive= df.Fare[(df.Survived==1) & (df.Pclass==z1)]
    ax.scatter(xlive, ylive, z1, zdir='z', color='pink',alpha=0.6)

for z2 in a:
    xdie= df.Age[(df.Survived==0) & (df.Pclass==z2)]
    ydie= df.Fare[(df.Survived==0) & (df.Pclass==z2)]
    ax.scatter(xdie, ydie, z2, zdir='z', color='red',alpha=0.6)

plt.show()
```

```python
#This part is describe figures for the relationship between age and fare in 3 Pclasses.
```
