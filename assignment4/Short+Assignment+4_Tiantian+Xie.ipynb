```python
def draw_pie(labels,quants):   
    pltb.figure(1, figsize=(6,6))   
    expl = [0,0.1]      
    colors  = ["red","blue"]     
    pltb.pie(quants, explode=expl, colors=colors, labels=labels, autopct='%1.1f%%',pctdistance=0.8, shadow=True)  
    pltb.title('Survival Rate', bbox={'facecolor':'0.7', 'pad':6})  

labels   = ['dead', 'survived']   
quants   = [549, 342]   
draw_pie(labels,quants)  

pltb.show()
```

___ For aged people (age>65) ___

```python
SeniorSurvived = df.loc[(df["Survived"] == 1) & (df["Age"] > 65)].Sex.count()
SeniorDead = df.loc[(df["Survived"] == 0) & (df["Age"] > 65)].Sex.count()
```

```python
# To show the survival rate in a pie chart

labels = 'Survived', 'Dead'
sizes = [SeniorSurvived, SeniorDead]
explode = (0, 0)

fig1, ax1 = pltb.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

pltb.show()
```

### 4.  Family Size

```python
##Definit the new columes "FamilySize"

df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df.tail(1)
```

```python
df.FamilySize.hist()
```

```python
pddf = pd.DataFrame()
pddf['PassengerNum'] = df[['FamilySize','Survived']].groupby(['FamilySize'])['Survived'].count()
pddf['FamilySize'] = pddf.index.values
pddf['Survived'] = df[['FamilySize','Survived']].groupby(['FamilySize'])['Survived'].sum()
# define x and y axises

fig = pltb.figure()
ax = fig.add_subplot(111)
pddf.Survived.plot(kind='line', color='red', ax=ax, label = 'Survivor', xticks = pddf['FamilySize'])
pddf.PassengerNum.plot(kind='line', color='grey', ax=ax, label = 'Passenger sum')
ax.legend()
pltb.title('Family Size: Survival', bbox={'facecolor':'1', 'pad':3})
# color the title

pltb.show()
```

** Summary: **
1. Most passengers travelled with small families around 1-4.
2. When less than 5 family members, the lager the family traveled together was, the more chance for them to survive. 
3. However, when the total number of family over 5, there was much less chance for them to survive.

### 5.  Fare & Family Size

This will be 2 dimentional chart - scatter, in order to explore the connection between fare and family size, when it comes to survival.

```python
fig, ax = pltb.subplots()

xalive = df.FamilySize[df.Survived==1]
xdead = df.FamilySize[df.Survived==0]
# define x axis
yalive = df.Fare[df.Survived==1]
ydead = df.Fare[df.Survived==0]
# define y axis
ax.set_xlabel('Family Size group')
ax.set_ylabel('Fare group')
m = ax.scatter(xalive, yalive, 8, 'red')
ax.hold
n = ax.scatter(xdead, ydead, 3, 'blue')
ax.hold
ax.legend([m,n],['Survivor','Victim'])
pltb.title('Family Size & Fare group', bbox={'facecolor':'0.8', 'pad':3}) 

pltb.show()
```

** Summary: **
1. The larger the family size was and the lower the fare was, the less opportunity for a passenger to survive.
2. Large family tended to buy cheaper ticket.
3. Passengers who bought more expensive ticket have more chance to survive.

### 6.  Age & Fare

### 7.  Age & Name Length & Pclass

To draw a 3 dimentional chart

```python
df['NameLength']=df.Name.apply(len)
# count the length of those names
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = pltb.figure()
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('Age group')
ax.set_ylabel('Name Length')
ax.set_zlabel('Pclass group')

a = df['Pclass']

for z1 in a:
    xlive= df.Age[(df.Survived==1) & (df.Pclass==z1)]
    ylive= df.NameLength[(df.Survived==1) & (df.Pclass==z1)]
    n = ax.scatter(xlive, ylive, z1, zdir='z', color='red', alpha=0.3, norm=0.5)

for z2 in a:
    xdead= df.Age[(df.Survived==0) & (df.Pclass==z2)]
    ydead= df.NameLength[(df.Survived==0) & (df.Pclass==z2)]
    m = ax.scatter(xdead, ydead, z2, zdir='z', color='blue', alpha=0.3, norm=0.5)

ax.legend([m,n],['Survivor','Victim'])

pltb.show()
```

** Summary: **
   
Short name, high class and young age contributed to more chance to live.
1. The longer the name was, the more likely for him or her to die.
2. Higher class meant more chance to be alive
3. Younger age meant more chance to be alive
