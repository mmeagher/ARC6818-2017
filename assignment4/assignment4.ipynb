```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('train.csv', header=0)
df
```

Overall survivial rate

```python
# Use Pie chart and histogram to analyse data 
sns.countplot(x='Survived',data=df)
sur=df['Survived'].sum()
Non_sur= 891-sur
labels = 'Survived','Non-survived'
sizes = [sur,Non_sur]
explode = [0.1,0] # highlight the survival rate
fig1,ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
ax1.axis('equal')
plt.show()
# Note here two diagrams are individual, the blue colour represents differently
```

Gender survival

```python
df.Sex.value_counts() #Count how many women and men abroad
```

```python
pd.crosstab(df['Sex'],df['Survived']).plot(kind='bar')
plt.show()
#The diagram shows that female's survival rate isbigger than males'. Guess people obeied 'Ladies first' during escaping.
```

Age survival

```python
# Overall age distribution
plt.figure(figsize=(15,5))
plt.subplot(121)
df['Age'].hist(bins = 60)
plt.xlabel('Age')
plt.ylabel('Number')
plt.show()

#Survival age distribution
plt.figure(figsize=(15,5))
plt.subplot(121)
df.Age[df.Survived==1].hist(bins = 60)
plt.xlabel('Age')
plt.ylabel('Number')
plt.show()

#Although some numbers are missing, still could find out that most survivors are among 20-35 years old.
```

Class  survival

```python
pd.crosstab(df['Pclass'],df['Survived']).plot(kind='bar')
plt.show()
#from the diagram could see the highest death rate is in Class3, the highest survival rate is in Class1. The reason might be
#ship staff let the rich people escape first.
```

Conclusion: Among 891 people in the ship, 38.4% survived. Survivors are mostly female, younger people(20-35 years), Class1(rich people).
