```python
import pandas as pd
import numpy as np
```

```python
Location = r'C:\Users\Admin\000Assignments\4\train.csv'
```

```python
df = pd.DataFrame(pd.read_csv(Location, header=0))
```

### In order to find which passenger attributes give the a good prediction of survival, relationships between the survival rate and other attributes need to be analysed.

### There are much missing cabin information, which not important for survival rate prediction, so we extract the cabin column from the database.

```python
df = df.drop(['Cabin'], axis=1)
df
```

## 1. Age

### In the orginal database, many age data are not available. We use median numbers here to fulfill missing age data.

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df
```

```python
# For aged people (age>65)

AgedSurvived = df.loc[(df["Survived"] == 1) & (df["Age"] > 65)].Sex.count()
AgedVictim = df.loc[(df["Survived"] == 0) & (df["Age"] > 65)].Sex.count()
```

```python
import matplotlib.pyplot as plt
```

```python
# To show the survival rate in a pie chart

labels = 'Survived', 'Not Survived'
sizes = [AgedSurvived, AgedVictim]
explode = (0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

```python
# For children (age<16)

ChildSurvived = df.loc[(df["Survived"] == 1) & (df["Age"] < 16)].Sex.count()
ChildVictim = df.loc[(df["Survived"] == 0) & (df["Age"] < 16)].Sex.count()
```

```python
labels = 'Survived', 'Not Survived'
sizes = [ChildSurvived, ChildVictim]
explode = (0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

```python
# For adults (age between 16 and 65)

AdultSurvived = df.loc[(df["Survived"] == 1) & (df["Age"][16:65])].Sex.count()
AdultVictim = df.loc[(df["Survived"] == 0) & (df["Age"][16:65])].Sex.count()
```

```python
labels = 'Survived', 'Not Survived'
sizes = [AdultSurvived, AdultVictim]
explode = (0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

### Pie charts above show that the survival rate for younger people was significantly higher.

## 2. Family Size

```python
# Create a new column to represent the family size.

df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df.FamilySize.describe()
```

```python
%matplotlib inline
```

```python
import pylab as P
df['FamilySize'].hist(bins=11, range=(1,11))
```

```python
pd.crosstab(df["FamilySize"],df["Survived"])
```

### Divide family sizes into three groups: Single (Family Size = 1), Small (Family Size from 2 to 4) and Large (Family Size larger than 4)

```python
# Set Single = 1, Small = 2, Large = 3

df.FamilySize[(df.FamilySize>1) & (df.FamilySize<=4)]=2
df
```

```python
df.FamilySize[df.FamilySize>4]=3
df
```

```python
pd.crosstab(df["FamilySize"],df["Survived"])
```

```python
pd.crosstab(df["FamilySize"],df["Survived"]).plot(kind="bar")
```

### The diagram above shows that small families were more possible to survive than single people and large families.

## 4. Class and fare

```python
# Firstly we use box plot to find the relationship between class and fare.

fig,ay = plt.subplots()
Farec1 = df.Fare[df.Pclass == 1]
Farec2 = df.Fare[df.Pclass == 2]
Farec3 = df.Fare[df.Pclass == 3]
plt.boxplot((Farec1,Farec2,Farec3),labels=("Pclass1","Pclass2","Pclass3"))
ay.set_ylim([-10,180])
ay.set_title("Boxplot of Fare and Pclass")
```

### The box plot shows that there is a obvious positive relationship between the class and fare, so in the following analysis we use class data only.

```python
pd.crosstab(df["Pclass"],df["Survived"])
```

```python
pd.crosstab(df["Pclass"],df["Survived"]).plot(kind="bar")
```

### The diagram above shows that passengers in the higher class level had higher suivival chance.

## Conclusion:
## From analysis above we could find that younger perople, females, people with small families and in higher classes were more likely to survive.
