```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

tdf=pd.read_csv('train.csv',header=0)
```

FamilySize

```python
SS = pd.DataFrame()
fig = plt.figure()
ax = fig.add_subplot(111)


tdf['FamilySize'] = tdf['SibSp'] + tdf['Parch'] + 1
SS['PassengerNum'] = tdf[['FamilySize','Survived']].groupby(['FamilySize'])['Survived'].count()
SS['FamSize'] = SS.index.values
SS['Survived'] = tdf[['FamilySize','Survived']].groupby(['FamilySize'])['Survived'].sum()
SS['Ratio'] = SS.Survived / SS.PassengerNum * 100

pd.crosstab(tdf["FamilySize"],tdf["Survived"]).plot(kind="line")
SS.Ratio.plot(kind='line', color='red', ax=ax, label = 'Ratio')
ax.set_ylabel('%')
ax.legend(['Ratio'])
plt.show()
```

```python
# The FamilySize under 4 has the lower death rate. On the other hand, those which are more than 4 have higher death rate.
```

Age & Fare & Pclass

```python
fig, ax = plt.subplots()

ylive = df.Fare[df.Survived==1]
ydie = df.Fare[df.Survived==0]
xlive = df.Age[df.Survived==1]
xdie = df.Age[df.Survived==0]

m = ax.scatter(xlive, ylive, 5, 'pink')
ax.hold
n = ax.scatter(xdie, ydie, 5, 'red')
ax.hold

ax.set_xlabel('Age')
ax.set_ylabel('Fare')

ax.legend([m,n],['Survived','Died'])

plt.show()
```

```python
# Age, Fare & Pclass

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('Age group')
ax.set_ylabel('Fare group')
ax.set_zlabel('Pclass group')

a = tdf['Pclass']

for z1 in a:
    xlive= tdf.Age[(tdf.Survived==1) & (tdf.Pclass==z1)]
    ylive= tdf.Fare[(tdf.Survived==1) & (tdf.Pclass==z1)]
    n = ax.scatter(xlive, ylive, z1, zdir='z', color='grey', alpha=0.3, norm=0.5)

for z2 in a:
    xdie= tdf.Age[(tdf.Survived==0) & (tdf.Pclass==z2)]
    ydie= tdf.Fare[(tdf.Survived==0) & (tdf.Pclass==z2)]
    m = ax.scatter(xdie, ydie, z2, zdir='z', color='pink', alpha=0.3, norm=0.5)

ax.legend([m,n],['Survivor','Victim'])

plt.show()
```

The price of tickets are focus on 0-100,the passengers' mostly from 15-50.And the dead rate is concentrite on this part,especially in 20-to-50 people.

```python
# From this graphic, passengers from lower fare , class has the most potencial to die. And most of the victims are from 10-50 age group.
```
