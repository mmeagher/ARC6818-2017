##Probabilities of survival of men, women and children in the Titanic according to their social condition an gender.

```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv',header=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df['familySize']= df['SibSp']+ df['Parch'] + 1

df['Gender']= df['Sex'].map ({'female':1,'male':0}).astype(int)
```

```python
# Create a scatterplot of,
            # Age in Pclass 1 as the x axis
plt.scatter(df['Age'][df['Pclass'] == 1], 
            # Fare in Pclass 1 as the y axis
            df['Fare'][df['Pclass'] == 1], 
            # the marker as
            marker='.', 
            # the color
            color='b',
            # the alpha
            alpha=0.5,
            # with size
            s = 100,
            # labelled this
            label='Pclass 1')

             # Age in Pclass 2 as the x axis
plt.scatter(df['Age'][df['Pclass'] == 2], 
            # Fare in Pclass 2 as the y axis
            df['Fare'][df['Pclass'] == 2], 
            # the marker as
            marker='.', 
            # the color
            color='r', 
            # the alpha
            alpha=0.5,
            # with size
            s = 100,
            # labelled this
            label='Pclass2')

            # Age in Pclass 3 as the x axis
plt.scatter(df['Age'][df['Pclass'] == 3], 
            # Fare in Pclass 3 as the y axis
            df['Fare'][df['Pclass'] == 3], 
            # the marker as
            marker='.', 
            # the color
            color='g', 
            # the alpha
            alpha=0.5, 
            # with size
            s = 100,
            # labelled this
            label='Pclass3')
# Chart title
plt.title('Age and Fare in relation to PClass')

# y label
plt.ylabel('Fare')

# x label
plt.xlabel('Age')

# and a legend
plt.legend(loc='upper right')

plt.show()
```

```python
df.Survived.value_counts()
```

```python
#retrieve from the original data set ones who survived and the ones who died

survived = df [df.Survived == 1]
died = df [df.Survived == 0]
```

```python
# Create a scatterplot of,
            # Age in Pclass 1 as the x axis
plt.scatter(survived ['Age'][survived['Pclass'] == 1], 
            # Fare in Pclass 1 as the y axis
            survived['Fare'][survived['Pclass'] == 1], 
            # the marker as
            marker='.', 
            # the color
            color='b',
            # the alpha
            alpha=0.5,
            # with size
            s = 100,
            # labelled this
            label='Pclass 1 survivors')

             # Age in Pclass 2 as the x axis
plt.scatter(survived['Age'][survived['Pclass'] == 2], 
            # Fare in Pclass 2 as the y axis
            survived['Fare'][survived['Pclass'] == 2], 
            # the marker as
            marker='.', 
            # the color
            color='r', 
            # the alpha
            alpha=0.5,
            # with size
            s = 100,
            # labelled this
            label='Pclass2 survivors')

            # Age in Pclass 3 as the x axis
plt.scatter(survived['Age'][survived['Pclass'] == 3], 
            # Fare in Pclass 3 as the y axis
            survived['Fare'][survived['Pclass'] == 3], 
            # the marker as
            marker='.', 
            # the color
            color='g', 
            # the alpha
            alpha=0.5, 
            # with size
            s = 100,
            # labelled this
            label='Pclass3 survivors')

# Chart title
plt.title('Age and Fare in relation to survivors by Pclass ')

# y label
plt.ylabel('Fare')

# x label
plt.xlabel('Age')

# and a legend
plt.legend(loc='upper right')

plt.show()
```

```python
#Retrieve survivors female
survivedF = survived [survived.Gender == 1]
#Retrieve survivors male
survivedM = survived [survived.Gender == 0]
#Retrieve died Female
diedF = died [died.Gender == 1]
#Retrieve died Male
diedM = died [died.Gender == 0]
```
