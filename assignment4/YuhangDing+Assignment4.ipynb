```python
pc1s = 136
pc1d = 80
pc2s = 87
pc2d = 97
pc3s = 119
pc3d = 372

pc1 = pc1s + pc1d
pc2 = pc2s + pc2d
pc3 = pc3s + pc3d

plt.figure(figsize = (6,6))
plt.pie([pc1,pc2,pc3],labels=['Pclass1','Pclass2','Pclass3'],autopct='%1.0f%%')
plt.title('Passemgers Class Number')

plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.pie([pc1s+pc2s+pc3s,pc1d+pc2d+pc3d],labels=['Survived','Died'],autopct='%1.0f%%')
plt.title('Survived and Died Passenmgers Number')

plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.pie([pc1s,pc1d],labels=['Survived','Died'],autopct='%1.0f%%')
plt.title('Survived and Died Class1 Passenmgers Number')

plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.pie([pc2s,pc2d],labels=['Survived','Died'],autopct='%1.0f%%')
plt.title('Survived and Died Class2 Passenmgers Number')

plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.pie([pc3s,pc3d],labels=['Survived','Died'],autopct='%1.0f%%')
plt.title('Survived and Died Class3 Passenmgers Number')

plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.pie([pc1s,pc2s,pc3s,pc1d,pc2d,pc3d],labels=['Survived Class1','Survived Class2','Survived Class3','Died Class1','Died Class2','Died Class3'],autopct='%1.0f%%')
plt.title('Survived and Died Class1,2,3 Passenmgers Number')

plt.show()
```
