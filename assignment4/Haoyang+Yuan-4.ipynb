```python
sur_num=df['Survived'].sum()
no_sur_num = 891 - sur_num

labels =  'Not Survived','Survived'
sizes = [no_sur_num,sur_num]

explode = (0, 0.2,) 
fig1, ax1 = plt.subplots()

ax1.pie(sizes,colors=['red','blue'], explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.title('Survival rate') 

plt.show()
```

```python
##In this 891 passengers, 38% and 62% survived and survived respectively
# As aboved show, the number of no-survived people is more than half of total population, overtake the the amount of survived passengers with 342 people.#
#So, in the next part, i will explore the main elements that could influence the result of survived sitution, especially age, sex, familysize, pclass, fare.# 
```

```python
## Age & Fare
```

```python
fig, ax = plt.subplots()

ylive = df.Fare[df.Survived==1]
ydie = df.Fare[df.Survived==0]
xlive = df.Age[df.Survived==1]
xdie = df.Age[df.Survived==0]

m = ax.scatter(xlive, ylive, 5, 'blue')
ax.hold
n = ax.scatter(xdie, ydie, 5, 'red')
ax.hold

ax.set_xlabel('Age')
ax.set_ylabel('Fare')

ax.legend([m,n],['Survived','Not Survived'])

plt.show()
```

```python
## As you can see, 
## According to the fare, people with higher fare are easier to survive. especially VIP.
## In age, baby and children have more opportunity to survive. and the death rate of young adults is  high.
```

```python
## Pclass
```

```python
pddf['PassengerNum'] = df[['Pclass','Survived']].groupby(['Pclass'])['Survived'].count()
pddf['Pclass'] = pddf.index.values
pddf['Survived'] = df[['Pclass','Survived']].groupby(['Pclass'])['Survived'].sum()

fig = plt.figure()
ax = fig.add_subplot(111)
pddf.Survived.plot(kind='line', color='blue', ax=ax, label = 'Survivors', xticks = pddf['Pclass'])
pddf.PassengerNum.plot(kind='line', color='grey', ax=ax, label = 'Num of Passengers')
ax.legend()
plt.title(' Survival Situation of different Pclass', bbox={'facecolor':'0.8', 'pad':3})

plt.show()
```

```python
## From this line chart, it is easy to find that people with lower Pclass are more difficult  to survive.
```

```python
# Familysize
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['Familysize'] = df['SibSp'] + df['Parch'] + 1
SS['PassengerNum'] = df[['Familysize','Survived']].groupby(['Familysize'])['Survived'].count()
SS['Famsize'] = SS.index.values
SS['Survived'] = df[['Familysize','Survived']].groupby(['Familysize'])['Survived'].sum()
SS['Ratio'] = SS.Survived / SS.PassengerNum * 100

pd.crosstab(df["Familysize"],df["Survived"]).plot(kind="line")
SS.Ratio.plot(kind='line', color='black', ax=ax, label = 'Ratio')
ax.set_ylabel('%')
ax.legend(['Ratio'])
plt.show()
```

```python
##As this chart show, the familysize with around 4 family members have a highest rate in surviving.
```
