```python
# this part aims to import the moudles i will use in this assignment,and load the csv file by pandas moudle to get the information i need.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv(r"e:/train.csv")
df
```

```python
# i want to figure out the connection between the survival and some factors that i get from the csv file like sex ,pclass,age group and so on.
# i notice that there are many missing values in the csv file so i should check the data before i start
columns_train=['Age','Sex','Pclass','Fare']
x=df[columns_train]
```

```python
x['Sex'].isnull().sum()
x['Pclass'].isnull().sum()
x['Fare'].isnull().sum()
x['Age'].isnull().sum()
#before analyzing the data, i found there are 177 missing values of age in the data
#i will fill the missing data with median of the existing data
x['Age']=x['Age'].fillna(x['Age'].median())
x['Age'].isnull().sum()
```

```python
#first,i want to foucs on the different survival rate between men and women by pie chart
mensurvivor= len(df.loc[(df['Survived']==1)&(df['Sex']=='male')])
menvictim= len(df.loc[(df['Survived']==0)&(df['Sex']=='male')])
womensurvivor= len(df.loc[(df['Survived']==1)&(df['Sex']=='female')])
womenvictim= len(df.loc[(df['Survived']==0)&(df['Sex']=='female')])

plt.figure(figsize= (10,5))
plt.subplot(121)
sizes = [mensurvivor, menvictim]
labels = ['survivor', 'victim']
explode = (0.1, 0)
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Men Survival rate')


plt.subplot(122)
explode = (0.1,0)
plt.pie([womensurvivor,womenvictim],explode=explode,labels=['survivor','victim'],autopct='%1.1f%%')
plt.title('Women Survival rate')


plt.show()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df
```

```python
#considering of the large amount of data about the age and the ticket fare of passengers,i choose the scatter diagram to analysis the data
figure, y = plt.subplots()
y.scatter(df.Fare[df.Survived==1],df.Age[df.Survived==1],7,'g')
y.scatter(df.Fare[df.Survived==0],df.Age[df.Survived==0],7,'r')

y.set_xlabel('Ticket fare')
y.set_ylabel('Age group')
y.legend(('survivor', 'victim'))
y.grid(True)
plt.show()
```

```python
# in this part i will analyze the connection between the survival rate and the  Pclass
# i import seaborn moudle to help me complete the data visualization
import seaborn as sns

plt.figure(figsize=(7,5))
sns.countplot(x='Pclass', data=df)
plt.title('People from different Pclass', bbox={'facecolor':'0.8', 'pad':3})

plt.show()
```

```python
pclass1survivor= len(df.loc[(df['Survived']==1)&(df['Pclass']==1)])
pclass1victim= len(df.loc[(df['Survived']==0)&(df['Pclass']==1)])
pclass2survivor= len(df.loc[(df['Survived']==1)&(df['Pclass']==2)])
pclass2victim= len(df.loc[(df['Survived']==0)&(df['Pclass']==2)])
pclass3survivor= len(df.loc[(df['Survived']==1)&(df['Pclass']==3)])
pclass3victim= len(df.loc[(df['Survived']==0)&(df['Pclass']==3)])

plt.figure(figsize= (10,3))

plt.subplot(131)
explode = (0.1,0)
plt.pie([pclass1survivor,pclass1victim],explode=explode,labels=['survivor','victim'],autopct='%1.1f%%')
plt.title('Pclass1 Survival rate')

plt.subplot(132)
explode = (0.1,0)
plt.pie([pclass2survivor,pclass2victim],explode=explode,labels=['survivor','victim'],autopct='%1.1f%%')
plt.title('Pclass2 Survival rate')

plt.subplot(133)
explode = (0.1,0)
plt.pie([pclass3survivor,pclass3victim],explode=explode,labels=['survivor','victim'],autopct='%1.1f%%')
plt.title('Pclass3 Survival rate')

plt.show()
```
