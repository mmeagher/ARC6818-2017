```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## open the file
df = pd.read_csv('train.csv', header=0)
df
```

```python
##survuved amount and rate


sns.countplot(x='Survived',data=df)
sur_num=df['Survived'].sum()
no_sur_num = 891 - sur_num
labels = 'Survived', 'Not Survived'
sizes = [no_sur_num,sur_num]
explode = (0, 0.1,)  # only "explode" the 2nd slice (i.e. 'Hogs')
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Survival Situation') 

plt.show()

```

```python
## different situation in different class

CLASS_1_SUR = df.loc[(df["Survived"] == 1) & (df["Pclass"] == 1) ].Sex.count()
CLASS_1_NOSUR = df.loc[(df["Survived"] == 0) & (df["Pclass"] == 1) ].Sex.count()
CLASS_2_SUR = df.loc[(df["Survived"] == 1) & (df["Pclass"] == 2) ].Sex.count()
CLASS_2_NOSUR = df.loc[(df["Survived"] == 0) & (df["Pclass"] == 2) ].Sex.count()
CLASS_3_SUR = df.loc[(df["Survived"] == 1) & (df["Pclass"] == 3) ].Sex.count()
CLASS_3_NOSUR = df.loc[(df["Survived"] == 0) & (df["Pclass"] == 3) ].Sex.count()

sns.countplot(x='Pclass',data=df)

labels = 'CLASS_1_SUR', 'CLASS_1_NOSUR'
sizes = [CLASS_1_SUR, CLASS_1_NOSUR]
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Class 1 Survival Situation') 
plt.show()

labels = 'CLASS_2_SUR', 'CLASS_2_NOSUR'
sizes = [CLASS_2_SUR, CLASS_2_NOSUR]
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Class 2 Survival Situation') 
plt.show()

labels = 'CLASS_3_SUR', 'CLASS_3_NOSUR'
sizes = [CLASS_3_SUR, CLASS_3_NOSUR]
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Class 3 Survival Situation') 
plt.show()


vals = np.array([[CLASS_1_SUR, CLASS_1_NOSUR], [CLASS_2_SUR, CLASS_2_NOSUR], [CLASS_3_SUR, CLASS_3_NOSUR]])
fig, ax = plt.subplots()
labels = 'CLASS_1_SUR', 'CLASS_1_NOSUR','CLASS_2_SUR', 'CLASS_2_NOSUR','CLASS_3_SUR', 'CLASS_3_NOSUR'
ax.pie(vals.flatten(), radius=1.2,
       colors=plt.rcParams["axes.prop_cycle"].by_key()["color"][:vals.shape[1]])
ax.pie(vals.sum(axis=1), radius=1)
ax.set(aspect="equal", title='ALL Class Survival Situation')

pddf = pd.DataFrame()
pddf['PassengerNum'] = df[['Pclass','Survived']].groupby(['Pclass'])['Survived'].count()
pddf['Pclass'] = pddf.index.values
pddf['Survived'] = df[['Pclass','Survived']].groupby(['Pclass'])['Survived'].sum()

fig = plt.figure()
ax = fig.add_subplot(111)
pddf.Survived.plot(kind='line', ax=ax, label = 'Survivor', xticks = pddf['Pclass'])
pddf.PassengerNum.plot(kind='line', ax=ax, label = 'Passenger sum')
ax.legend()
plt.title('Class group: Survival', bbox={'facecolor':'0.8', 'pad':3})

plt.show()
```

```python
##Sex

pd.crosstab(df["Sex"],df["Survived"]).plot(kind="bar")
```

```python
##familysize


df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
pd.crosstab(df["FamilySize"],df["Survived"]).plot(kind="bar")
```

```python
##Embarked
sns.factorplot('Embarked','Survived', data=df,size=4,aspect=2)    

plt.show()
```

```python
##Age 
fig, ax = plt.subplots()

x1 = df.Age[df.Survived==1]
x2 = df.Age[df.Survived==0]
y1 = df.Fare[df.Survived==1]
y2 = df.Fare[df.Survived==0]
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
m = ax.scatter(x1, y1, 8)
ax.hold
n = ax.scatter(x2, y2, 3)
ax.hold
ax.legend([m,n],['Survived', 'Not Survived'])
plt.title('Age Survival Situation', bbox={'facecolor':'0.8', 'pad':3}) 

plt.show()
```
