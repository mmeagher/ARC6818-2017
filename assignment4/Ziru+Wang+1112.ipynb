```python
survived_num = df['Survived'].sum()
died_num = 891 - survived_num

plt.pie([died_num, survived_num],colors=['yellow','grey'], labels=['Survived','died'],autopct='%1.0f%%')
plt.title('Survival rate') 
plt.show()
df.Survived.value_counts()
```

```python
fig, ax = plt.subplots()

x_survive = df.Pclass[df.Survived==1]
x_died = df.Pclass[df.Survived==0]
y_survive = df.Age[df.Survived==1]
y_died = df.Age[df.Survived==0]

m = ax.scatter(x_survive, y_survive, 20, 'yellow')
ax.hold
n = ax.scatter(x_died, y_died, 5, 'grey')
ax.hold

ax.set_xlabel('Pclass')
ax.set_ylabel('Passenger Age')

ax.legend(['Survived','Death'])
ax.grid(True)
plt.show()
```
