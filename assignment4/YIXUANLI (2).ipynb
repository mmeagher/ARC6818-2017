# 1.Pclass
FIRSTLY,I ANALYSIS THE INFLUENCE OF PASSENGERS'S CLASS ON SURVIVAL.

```python
pd.crosstab(data.Survived, data.Pclass).plot(kind='bar')
```

ACCORDING TO THE CHART ABOVE, I ANALYSIS THE PERCENT OF DIFFERENT CLASSES PASSENGERS WHO ARE DEAD.

```python
HIGH = data.loc[(data["Pclass"] == 1) & (data['Survived']==0)].Sex.count()
MIDDLE = data.loc[(data["Pclass"] == 2) & (data['Survived']==0)].Sex.count()
LOW = data.loc[(data["Pclass"] == 3) & (data['Survived']==0)].Sex.count()

labels = 'HIGH', 'MIDDLE', 'LOW'
sizes = [HIGH, MIDDLE,LOW]
explode = (0.05,0.05,0.05)
fig1, ax1 = chris.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.0f%%')
ax1.axis('equal')


chris.show()
```

THIS CHART SHOWS THAT LOW CLASS PASSENGERS ARE MORE LIKELY TO BE DEAD.

# 2.SEX
I ANALYSIS DIFFERENT SEX FOR SURVIVAL. 

```python
pd.crosstab(data.Sex, data.Survived).plot(kind="bar")
```

IT SHOWS THAT AGE,PCLASS,SIBSP,PARCH,TICKET,FARE AND CABIN MAY HAVE SOME CONNECT WITH SURVIVED.

BUT THERE ARE SOME MISSING STATISTICS OF AGE AND CABIN.

FISTLY, I WILL ANALYSIS THE COUNT OF PEOPLE WHO SURVIVED OR NOT. 

```python
AGELIVE = data.loc[(data["Survived"] == 1)].Sex.count()
AGEDEAD = data.loc[(data["Survived"] == 0)].Sex.count()

labels = 'LIVE', 'DEAD'
sizes = [AGELIVE, AGEDEAD]
explode = (0.05, 0.05)
fig1, ax1 = chris.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.0f%%')
ax1.axis('equal')
chris.show()
```

COMBINE SIBSP WITH PARCH AS FAMILYSIZE TO ANALYSIS ITS INFLUENCE ON SURVIVAL.
