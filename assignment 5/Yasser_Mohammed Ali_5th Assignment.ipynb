```python
footbplrs = FF[['Name', 'Age', 'Club', 'Overall']]
```

```python
footbplrs.isnull().head(10)

# Getting rid of the NaN values
```

```python
footbplrs.isnull().sum()
```

```python
footbplrs = footbplrs.dropna()
```

```python
footbplrs.sort_values('Age').head(20)

# Listing players from the oldest to the youngest player
```

```python
footbplrs.sort_values('Age', ascending=False).head(20)

# Listing players from oldest to youngest player
```

```python
footbplrs.groupby('Club').mean().sort_values('Age')

# Finding out the average players age for each club
```

```python
RealMadrid_footbplrs = footbplrs[footbplrs['Club'] == 'Real Madrid CF']

# Extracting Real Madrid (my favourite team) players from the whole list  
```

```python
RealMadrid_footbplrs.sort_values('Age', ascending=False)

# Sorting Real Madrid players gradually by age (from oldest to youngest)
```

```python
footbplrs= FF[['Name', 'Age', 'Club', 'Overall', 'Wage']].dropna()
```

```python
footbplrs.head(10)
```

```python
footbplrs.dtypes

# Checking the value type of "Wage" so we can list them in an ascending or descending order
```

```python
wages = footbplrs['Wage'].map(lambda x: x.replace('‚Ç¨', ''))

# Getting rid of the "‚Ç¨" symbol in the "Wage" column
```

```python
wages.head()
```

```python
wages = wages.map(lambda x: x.replace('K', ''))

# Getting rid of the "‚Ç¨" symbol in the "Wage" column
```

```python
wages = wages.astype('int')*1000

# Multiplying each "Wage" value by 1000 instead of the letter "K"
```

```python
footbplrs.loc[:, 'Wage'] = wages.values

# Assigning the new values to the "Wage" column
```

```python
footbplrs.dtypes

# And Wage values are now converted from "object" to "integer"
```

```python
footbplrs.groupby('Club')['Wage'].mean().sort_values(ascending= False).head(10)

# Finding out which team spends most on his players on average
```

```python
footbplrs.groupby('Club').mean().sort_values('Wage', ascending = False).head(10)

# Finding out the exact Wage each team players receive from their teams, in a descending order
```

```python
footbplrs.sort_values('Wage', ascending = False).head(20)

# Listing players from the one with the highest wage to the lowest wage
```

```python
footbplrs = FF [['Name', 'Age', 'Club', 'Overall', 'Nationality']].dropna()
```

```python
% matplotlib inline
```

```python
footbplrs.groupby('Nationality').Name.count().sort_values(ascending= False).head(10).plot(kind='barh')

# to know how many players each country has in FIFA 18 in a horizontal bar chart
```

```python
footbplrs.groupby('Nationality').Name.count()

# to know how many players each country has in FIFA 18
```

```python
footbplrs.groupby('Nationality').Name.count().sort_values(ascending = False).head(10)

# to know how many players each country has in FIFA 18 but now in a descending order
```

```python
footbplrs.groupby('Nationality').Name.count().sort_values(ascending = False).head(10).plot(kind = 'pie')

# # to know how many players each country has in FIFA 18 using pie chart
```
