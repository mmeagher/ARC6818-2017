### A further study for the distribution of different noise level in decibel (dB) from road sources is carried out as follows since road sources occupies the largest percentage of total noise exposure.

### iii. Noise Level by Road Sources vs Location

### From the information of the dataset, the noise exposure level was measured in two ways, Lden and Lnight. Lden is the average noise level for day, evening and night thus Lden data will be used for the analysis.

```python
# These are the data of Lden noise level for road sources.
Road1=df['Road_Pop_Lden>=55dB']
Road2=df['Road_Pop_Lden>=60dB']
Road3=df['Road_Pop_Lden>=65dB']
Road4=df['Road_Pop_Lden>=70dB']
Road5=df['Road_Pop_Lden>=75dB']
```

```python
# This is to plot a stacked bar chart for noise exposure level distribution for location.
plotly.tools.set_credentials_file(username='skylar24', api_key='vRRP9Dqpd9e5NJu7uE9V')

trace1 = go.Bar(y=(loc),x=(Road1),name='Road>=55dB',orientation = 'h' )
trace2 = go.Bar(y=(loc),x=(Road2),name='Road>=60dB',orientation = 'h' )
trace3 = go.Bar(y=(loc),x=(Road3),name='Road>=65dB',orientation = 'h' )
trace4 = go.Bar(y=(loc),x=(Road4),name='Road>=70dB',orientation = 'h' )
trace5 = go.Bar(y=(loc),x=(Road5),name='Road>=75dB',orientation = 'h' )

data = [trace1, trace2, trace3, trace4, trace5]
layout = go.Layout(barmode='stack', width=1000, height=1200, title='Noise Exposure Distribution by Lden',
                   font=dict(size=7))
                
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

### From above, it shows that noise exposure level between 55dB and 60dB occupies approximately half of the total noise exposure. In addition, noise exposure between 60dB and 65dB is the second large exposure level and then range between 65dB and 70dB is the third large level range. These three range occupies almost all the noise exposure. Furthermore, the maximum safe sound level limit is 85dB, thus these noise exposure is relatively unharmful for people. Also, since the lowest noise exposure level range, 55dB to 60dB has about half occupancy, the environment for people in UK can be seen as healthy and comfortable.
