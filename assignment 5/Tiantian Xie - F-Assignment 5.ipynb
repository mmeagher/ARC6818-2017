### 1 Chart-Weekly Working Hours

#### Australia

```python
fig = plt.figure(figsize=(10,6))
# change the figure size
ax = fig.add_subplot(111)
df_1.WeeklyWorkingHours.head(13).plot(kind="bar", color="Red")
# set the information for the chart
plt.xlabel('Year 1870-2000')
plt.ylabel('Working Hours/week')
# set labels for x&y axis
plt.title('Australia')
# title of the chart
```

The figure dropped from 56 in 1870 to 40 in 1950, and then it became stable in the remainning 60 years.

#### Belguim

```python
fig = plt.figure(figsize=(10,7))
# change the figure size
ax = fig.add_subplot(111)
df_1.WeeklyWorkingHours[13:24].plot(kind="bar", color="orange")
# set the information for the chart
plt.xlabel('Year 1870-2000')
plt.ylabel('Working Hours/week')
# set labels for x&y axis
plt.title('Belgium')
# title of the chart
```

The figure experienced a steady drop by 30 hours from 70 over these 13 decades.

#### Canada

```python
fig = plt.figure(figsize=(10,6))
# change the figure size
ax = fig.add_subplot(111)
df_1.WeeklyWorkingHours[25:36].plot(kind="bar", color="yellow")
# set the information for the chart
plt.xlabel('Year 1870-2000')
plt.ylabel('Working Hours/week')
# set labels for x&y axis
plt.title('Canada')
# title of the chart
```

The figure rose from 57 hours per week (1870) to 60 hours at the beginning.
After that it sharply decreased to less than 40 hours in 2000.

#### Switzerland

```python
fig = plt.figure(figsize=(10,7))
# change the figure size
ax = fig.add_subplot(111)
df_1.WeeklyWorkingHours[136:147].plot(kind="bar", color="green")
# set the information for the chart
plt.xlabel('Year 1870-2000')
plt.ylabel('Working Hours/week')
# set labels for x&y axis
plt.title('Switzerland')
# title of the chart
```

Weekly working hours smoothly dropped over this 13 decades.

#### United Kingdom

```python
fig = plt.figure(figsize=(10,6))
# change the figure size
ax = fig.add_subplot(111)
df_1.WeeklyWorkingHours[147:172].plot(kind="bar", color="blue")
# set the information for the chart
plt.xlabel('Year 1870-2000')
plt.ylabel('Working Hours/week')
# set labels for x&y axis
plt.title('United Kingdom')
# title of the chart
```

Despite the fluctuation in the meddle of 1870-2000 period, the weekly working hours decreased by 18 hours over the 13 decades.

#### Summary
1. This data – shown in the following visualization – shows that over this time working hours have steeply declined. Full-time workers in these countries work 20 or even 30 hours less every week than in the 19th century.
2. In 2000, as for those five countries analysed, the working hours per week seem to be almost all around 40 hours.
