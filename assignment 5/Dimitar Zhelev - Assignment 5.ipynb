#### 3.1.1 During the proccess of work, untill the final visualization 'HappinessRank' and 'Generocity' will be taken as the variables

```python
from bokeh.models.widgets import Panel, Tabs
# importing panels will combine 2 types of visualization in one by creating 'Tabs'
from bokeh.plotting import figure

p1 = figure(title=('Data for 2015'),
            height=400, width=800)
# adjusting the height and width is important so the data is presented in a readable way
# 'height', 'width', 'range', 'axis' can be used to resize the canvas
p1.circle(x=af.loc[10:100].HappinessRank, y=af.loc[10:100].Generosity,
          size=15, fill_color="orange", line_color="green", line_width=2, alpha=0.5)
tab1 = Panel(child=p1, title="Circle H/G")

p2 = figure(height=400, width=800)
p2.line(x=af.loc[10:100].HappinessRank, y=af.loc[10:100].Generosity,
        line_width=3, color="navy", alpha=0.5)
tab2 = Panel(child=p2, title="Line H/G")

# H/G = HappinessRank / Generocity
# line is added for a better visualization of the order in the visualization

p1.xaxis[0].formatter = NumeralTickFormatter(format="0")
p1.yaxis[0].formatter = NumeralTickFormatter(format="%0.0")

p2.xaxis[0].formatter = NumeralTickFormatter(format="0")
p2.yaxis[0].formatter = NumeralTickFormatter(format="%0.0")

# X axis is base on rank so it is a round number
# Y axis is based on '0.000...' type of numbers, therefore, I use 0.0 formating method for the numbers

p1.xaxis.axis_label = "HappinessRank"
p1.yaxis.axis_label = "Generosity"

p2.xaxis.axis_label = "HappinessRank"
p2.yaxis.axis_label = "Generosity"

tabs = Tabs(tabs=[ tab1, tab2 ])

show(tabs)
```

#### 3.2 Creating my own 'Dictionary' to target data within a specific range from the whole data source

```python
from bokeh.models import ColumnDataSource
source = ColumnDataSource(dict(
    x=af.loc[10:100].HappinessRank,
    y=af.loc[10:100].Generosity,
    country=af.loc[10:100].Country,
    ))

# For this visualization I have picked only 3 columns
```

```python
# source.data
# with the 'source.data' we can call the full dictionary

source.column_names
# to make sure of the names of the columns that the data is storing
```

#### 3.3 Starting to insert interactive tools

```python
from bokeh.models import HoverTool
# inserting the Hover tool
```

```python
hover = HoverTool(tooltips='@country')
# @ is added so the hover tool knows which country to pick from the dictionary

plot = dict(height=600, width=900, x_axis_type='log', x_range=(10,140), y_range=(-0.05,0.7))
# height, width are needed to resize the graph, but if 'x_range' is not aded then the big circles will be partly
# out of the vview range. That is why i am adding all of them, to be able to control the canvas fully

p = figure(title=('Data for 2015'),
           tools=[hover],**plot)

# tools is adding another tool to our visualization

p.circle(x='x', y='y',
         fill_color="yellow", line_color="black", line_width=2,
         size='x', alpha=0.6, source=source)

# the size is according to the 'X' column, which is the HappinessRank
# in the previous graph the size was a fixed number, but now I am using a column data form my dictionary
# the most efficient one for presenting would be the 'Happiness Rank'

p.xaxis.axis_label = "HappinessRank"
p.yaxis.axis_label = "Generosity"

# labeling the 'X' and 'Y' axis to show what are they presenting

show(p)

# This is exactly the same figure but the data has been taken from the 'Dictionary' we have manually created
# The relationship between the placement of the circles and its rank has also been visualizad
# The smalled th rank number is the better = resulting in smaller circle size
# The rank is going down from left to right as visible on the graph

# HOVER - use the mouse to hover over the circels and see which country is the circle you are targeting
```

#### 3.4 Using the whole data souce

```python
source = ColumnDataSource(af)
source.column_names

# Untill now I explored how I can take specific number of rows and columns and manually create a dictionary
# where from now on I will use the CSV file full data

# To do so first I will also remove the df.loc[] by using only df -> this will take all the rows data
# I am also removing the 'dict()' and taking all the columns data as well

# this is also a shorter way of creating dictionary
# both ways are good and have pros and cons
```

```python
plot = dict(height=600, width=900)

p = figure(title=('Data for 2015'),
           **plot)

p.circle(x='HappinessRank', y='Generosity',
         fill_color="yellow",
         line_color="black",
         line_width=2,
         size='Population',
         alpha=0.6, source=source)

# I am adding another variable by changing the size to a different column data, to the Population
# 'alpha' is for transperancy
# By changing the size according to the data, the size becomes too big 
# The data is too big to be presented in the canvas of the screen
# In the next step i will need to use a way of resizing the data

p.xaxis.axis_label = "HappinessRank"
p.yaxis.axis_label = "Generosity"

show(p)

# Big data has to be managed properly on the presented canvas, otherwise it will happen this - The circles are too big now
```

```python
from bokeh.models import LinearInterpolator

size_mapper = LinearInterpolator(
    x=[af.Population.min(), af.Population.max()],
    y=[10,100]
)

# adding size_mapper I can achieve a better overview if I dont want to use zoom
# size mapper can controll the collected data, which determines the size
# in other words, I add a variable and can controll the size that is according to a fixed number

hover = HoverTool(tooltips='@Country,@Population')
# Specifying what the hover tool to show 

p = figure(
    title=('Data for 2015'),
    toolbar_location='above',
    tools=[hover], **plot)


p.circle(x='HappinessRank', y='Generosity',
         fill_color="yellow",
         line_color="black",
         line_width=2,
         size={'field': 'Population', 'transform': size_mapper},
         # giving viriables to the 'size_mapper'
         alpha=0.6, source=source)

p.xaxis.axis_label = "HappinessRank"
p.yaxis.axis_label = "Generosity"

show(p)

# The Hover is presenting the Country and the Population when the mouse is on top of any circle
```

#### 3.5 Dividing the MASS into GROUPS

```python
# In the next step I would like to divide the big mass of circles into groups
# To do that I will group them by their 'Region' data from the table and color them with a unique collor
```

```python
list(af.Region.unique())
# using only the unique names of the Region list will give me only one name of each region,
# where each one will be the name of each new group
```

```python
from bokeh.palettes import Spectral10
Spectral10

# importing collor pallets
# going to bokeh.pydata.org can be useful to chose a type of pallets that will suit the type of presented data
```

```python
from bokeh.models import LinearInterpolator, CategoricalColorMapper


size_mapper = LinearInterpolator(
    x=[af.Population.min(), af.Population.max()],
    y=[10,100]
)
# the canvas size is set by the LinearInerpolator now and 'X' and 'Y' = to their given variables

color_mapper = CategoricalColorMapper(
    factors=list(af.Region.unique()),
    palette=Spectral10,
)

# defining the relation between the color and the created list of Region groups



hover = HoverTool(tooltips=('@HappinessRank,@Country,@Region'))
# in the previous I presented the Country and the Population data, but now I want to show the Rank/Country and Region


p = figure(
    title=('Data for 2015'),
    toolbar_location='above',
    tools=[hover], **plot)

p.circle(x='HappinessRank', y='Generosity',
         line_color="black",
         line_width=1,
         size={'field': 'Population', 'transform': size_mapper},
         color={'field': 'Region', 'transform': color_mapper},
         # giving viriables to the 'color_mapper'
         alpha=0.6, source=source,
        legend='Region',
         #importing a legend for the Groups presented by each color
)

p.legend.border_line_color = None
p.right = p.legend
# to replace the legend and remove it from the graph canvas I am using 'p.right' specifically for the legend

p.xaxis.axis_label = "HappinessRank"
p.yaxis.axis_label = "Generosity"

show(p)
```

### 4. Visualizing and Comparing data for the years of 2015-2016-2017

```python
# First I will be creating the lists
```

```python
list(af.Region.unique())
list(bf.Region.unique())
list(cf.Region.unique())
```

```python
# Now it is time to combine all the 3 CSV files and the used methods to present and compare the data
```

```python
source0 = ColumnDataSource(af)
source1 = ColumnDataSource(bf)
source2 = ColumnDataSource(cf)

from bokeh.models.widgets import Panel, Tabs
from bokeh.io import output_file, show

from bokeh.models import LinearInterpolator, CategoricalColorMapper


size_mapper = LinearInterpolator(
    x=[af.Population.min(), af.Population.max()],
    y=[10,100]
)

size_mapper = LinearInterpolator(
    x=[bf.Population.min(), bf.Population.max()],
    y=[10,100]
)

size_mapper = LinearInterpolator(
    x=[cf.Population.min(), cf.Population.max()],
    y=[10,100]
)


color_mapper = CategoricalColorMapper(
    factors=list(af.Region.unique()),
    palette=Spectral10,
)

color_mapper = CategoricalColorMapper(
    factors=list(bf.Region.unique()),
    palette=Spectral10,
)

color_mapper = CategoricalColorMapper(
    factors=list(cf.Region.unique()),
    palette=Spectral10,
)

hover = HoverTool(tooltips=('@HappinessRank,@Country,@Region'))

p = figure(
    title=('2015,2016,2017 Happiness and Generosity'),
    toolbar_location='above',
    tools=[hover],**plot)

invisible_circle0 = p.circle(x='HappinessRank', y='Generosity',
         line_color="black",
         line_width=1,
         size={'field': 'Population', 'transform': size_mapper},
         color={'field': 'Region', 'transform': color_mapper},
         alpha=0.6, source=source0,
        legend='Region')
        
invisible_line0 = p.line(x='HappinessRank', y='Generosity',
        line_width=3, color="navy", alpha=0.5,source=source0)


invisible_circle1 = p.circle(x='HappinessRank', y='Generosity',
         line_color="black",
         line_width=1,
         size={'field': 'Population', 'transform': size_mapper},
         color={'field': 'Region', 'transform': color_mapper},
         alpha=0.6, source=source1,
        )

invisible_line1 = p.line(x='HappinessRank', y='Generosity',
        line_width=3, color="red", alpha=0.5,source=source1)


invisible_circle2 = p.circle(x='HappinessRank', y='Generosity',
         line_color="black",
         line_width=1,
         size={'field': 'Population', 'transform': size_mapper},
         color={'field': 'Region', 'transform': color_mapper},
         alpha=0.6, source=source2,
        )

invisible_line2 = p.line(x='HappinessRank', y='Generosity',
        line_width=3, color="green", alpha=0.5,source=source2)


invisible_circle0.visible = False
invisible_circle1.visible = False
invisible_circle2.visible = False

invisible_line0.visible = False
invisible_line1.visible = False
invisible_line2.visible = False


code = '''\
object.visible = toggle.active
'''

# With the following I am setting the Toggles to refer to a different graph


callback0 = CustomJS.from_coffeescript(code=code, args={})
toggle0 = Toggle(label="Circle 2015", button_type="success", callback=callback0)
callback0.args = {'toggle': toggle0, 'object': invisible_circle0}

callback1 = CustomJS.from_coffeescript(code=code, args={})
toggle1 = Toggle(label="Circle 2016", button_type="success", callback=callback1)
callback1.args = {'toggle': toggle1, 'object': invisible_circle1}

callback2 = CustomJS.from_coffeescript(code=code, args={})
toggle2 = Toggle(label="Circle 2017", button_type="success", callback=callback2)
callback2.args = {'toggle': toggle2, 'object': invisible_circle2}

callback3 = CustomJS.from_coffeescript(code=code, args={})
toggle3 = Toggle(label="Line 2015", button_type="success", callback=callback3)
callback3.args = {'toggle': toggle3, 'object': invisible_line0}

callback4 = CustomJS.from_coffeescript(code=code, args={})
toggle4 = Toggle(label="Line 2016", button_type="success", callback=callback4)
callback4.args = {'toggle': toggle4, 'object': invisible_line1}

callback5 = CustomJS.from_coffeescript(code=code, args={})
toggle5 = Toggle(label="Line2017", button_type="success", callback=callback5)
callback5.args = {'toggle': toggle5, 'object': invisible_line2}


p.legend.border_line_color = None
p.right = p.legend



p.xaxis.axis_label = "HappinessRank"
p.yaxis.axis_label = "Generosity"

show(layout([p],[toggle0,toggle1,toggle2,toggle3,toggle4,toggle5]))

# By clicking the buttons we can compare the years and if clicking multiple buttons we can overlay the graph
# and compare different years on one canvas

# The graphs can be combines Circles + Circles or Circles + Lines or Lines + Lines to explore the changes
```
