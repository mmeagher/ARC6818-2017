

#  Explore movie Ratings Visualization 

Dataset:Movie-Ratings.csv



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
mv = pd.read_csv("Movie-Ratings.csv")
```

```python
mv
```




## reorganize the columns to make it clearer

```python
mv.columns = ['film','genre','critics_rating','audience_rating','budget','yor']
```

```python
mv
```



## explore the relationship between movie critics rating,audience rating and movie genres

```python
sns.boxplot(x = mv.genre, y = mv.critics_rating, )
```

```python
sns.violinplot(x = mv.genre, y = mv.audience_rating, gridsize=500,width=1)
```

```python
sns.distplot(mv.budget, color = 'red')
```



## the rating of defference genres 

```python
sns.lmplot(data=mv, x='critics_rating', y='audience_rating', hue='genre', size=6, fit_reg = False, aspect=2, )
```



## the frequency of critics rating

```python
plt.hist(mv.critics_rating, color='red',label='budget')
plt.xlabel('critics_rating')
plt.ylabel('frequency')
```



## the relationship of different variables

```python
f, axes = plt.subplots(3,2, figsize = (16,14))
k0 = sns.kdeplot(mv.budget, mv.audience_rating,shade = True, cmap="plasma",ax = axes[0][0])
k1 = sns.kdeplot(mv.audience_rating, mv.critics_rating,shade = True, cmap = "spring", ax = axes[0][1])
k2 = sns.kdeplot(mv.budget, mv.critics_rating, shade = True, cmap='rainbow', ax= axes[1][0])
k3 = sns.kdeplot(mv.yor, mv.budget, shade = True, cmap='cool', gridsize =300,ax= axes[1][1])
k3 = sns.kdeplot(mv.yor, mv.audience_rating, shade = True, cmap='copper', gridsize =300,ax= axes[2][0])
k3 = sns.kdeplot(mv.yor, mv.critics_rating, shade = True, cmap='Spectral', gridsize =300,ax= axes[2][1])
```

```python
g = sns.FacetGrid(data= mv,row = 'genre', col = 'yor', hue="genre")
kws = dict(s=50,linewidth=2, edgecolor="black")
g = g.map(plt.scatter, 'critics_rating', 'audience_rating')
```

```python
g = sns.FacetGrid(data= mv,row = 'genre', col = 'yor', hue="genre")
g = g.map(plt.hist,'audience_rating')
```
